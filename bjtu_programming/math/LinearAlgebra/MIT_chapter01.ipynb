## 方程组的几何解释
### 二元一次方程组
$$
\begin{cases}
x&-2y&=1\\
3x&+2y&=11
\end{cases}
，矩阵形式为
\begin{bmatrix}
1 & -2 \\
3 & 2
\end{bmatrix}
\begin{bmatrix}
x\\y
\end{bmatrix}
=
\begin{bmatrix}
1\\11
\end{bmatrix}
$$
#### 行图像
每个方程在平面上的图像为$xoy$平面上的一条直线,两直线交点$(3,1)$即为方程组的解。</br>
<img src="pictures/2_1.png" title="2_1"  width=500 heigth=400/>

#### 列图像
原方程组的求解变为了寻找满足如下要求的$x$和$y$，使得系数矩阵列向量的**线性组合**等于向量 $b$。
$$
x\begin{bmatrix} 1 \\ 3  \end{bmatrix}
+ y\begin{bmatrix} -2 \\ 2 \end{bmatrix}
= \begin{bmatrix} 1 \\ 11 \end{bmatrix}
=b
$$
代入交点$x=3,y=1$，可得
$$
3\begin{bmatrix} 1 \\ 3  \end{bmatrix}
+ \begin{bmatrix} -2 \\ 2  \end{bmatrix}
= \begin{bmatrix} 1 \\ 11  \end{bmatrix}
$$
其几何图像如下图。

### 三元一次方程组
$$
\begin{cases}
x&+2y&+3z&=6\\
2x&+5y&+2z&=4\\
6x&-3y&+z&=2
\end{cases}
，矩阵形式为
Ax=b,即 \begin{bmatrix}
1 & 2 & 3 \\
2 & 5 & 2 \\
6 & -3 & 1
\end{bmatrix}
\begin{bmatrix}
x\\y \\z
\end{bmatrix}
=
\begin{bmatrix}
6\\4\\2
\end{bmatrix}
$$
#### 行图像
每个方程在3维空间的图像为一个平面,方程组的解为多个平面的交集(可能为点、线、面)。</br>
<img src="pictures/2_3.png" width=400 heigth=300/>

#### 列图像
原方程组的求解变为了寻找满足如下要求的$x$、$y$和$z$，使得系数矩阵列向量的**线性组合**等于向量 $b$。
$$
x\begin{bmatrix} 1 \\ 2 \\ 6 \end{bmatrix}
+ y\begin{bmatrix} 2 \\ 5 \\ -3 \end{bmatrix}
+ z\begin{bmatrix} 3 \\ 2 \\ 1\end{bmatrix}
= \begin{bmatrix}
6\\4\\2
\end{bmatrix}
=b
$$
代入交点$(x,y,z)=(0,0,2)$，可得
$$
0\begin{bmatrix} 1 \\ 2 \\ 6 \end{bmatrix}
+ 0\begin{bmatrix} 2 \\ 5 \\ -3 \end{bmatrix}
+ 2\begin{bmatrix} 3 \\ 2 \\ 1\end{bmatrix}
= \begin{bmatrix}
6\\4\\2
\end{bmatrix}
$$
其几何图像如下图。

## 矩阵消元
### 消元法 Method of Elimination
高斯消元法（Gauss elimination）就是通过对方程组中的某两个方程进行适当的数乘和加，以达到将某一未知数系数变为零，从而削减未知数个数的目的。
有三元方程组
$$
\begin{cases}x&+2y&+z&=2\\3x&+8y&+z&=12\\&4y&+z&=2\end{cases}，
对应的矩阵形式为 
Ax=b,即
\begin{bmatrix}1&2&1\\3&8&1\\0&4&1\end{bmatrix}\begin{bmatrix}x\\y\\z\end{bmatrix}=\begin{bmatrix}2\\12\\2
\end{bmatrix}
$$

* `消元`（操作系数矩阵$A$分别消去$x、y$）  
$
\begin{bmatrix}\underline{1}&2&1\\3&8&1\\0&4&1\end{bmatrix}\xrightarrow{row_2-3row_1}\begin{bmatrix}\underline{1}&2&1\\0&2&-2\\0&4&1\end{bmatrix}
\xrightarrow{row_3-2row_2}\begin{bmatrix}\underline{1}&2&1\\0&\underline{2}&-2\\0&0&\underline{5}\end{bmatrix}
$

* `回代`（增广矩阵）  
$
\left[\begin{array}{c|c}A&b\end{array}\right]=\left[\begin{array}{ccc|c}1&2&1&2\\3&8&1&12\\0&4&1&2\end{array}\right]\to\left[\begin{array}{ccc|c}1&2&1&2\\0&2&-2&6\\0&4&1&2\end{array}\right]\to\left[\begin{array}{ccc|c}1&2&1&2\\0&2&-2&6\\0&0&5&-10\end{array}\right]
$  
此时方程组为  
$
\begin{cases}x&+2y&+z&=2\\&2y&-2z&=6\\&&5z&=-10\end{cases}
\implies
\begin{cases}x&=2\\y&=1\\z&=-2\end{cases}
$

### 消元矩阵
* 消元法第1步操作  
$
E_{2,1}A
=\begin{bmatrix}1&0&0\\-3&1&0\\0&0&1\end{bmatrix}\begin{bmatrix}\beta_1\\ \beta_2 \\ \beta_3\end{bmatrix}
=\begin{bmatrix}\beta_1\\ \beta_2-3\beta_1 \\ \beta_3\end{bmatrix}
$
，即
$
\begin{bmatrix}1&0&0\\-3&1&0\\0&0&1\end{bmatrix}\begin{bmatrix}1&2&1\\3&8&1\\0&4&1\end{bmatrix}=\begin{bmatrix}1&2&1\\0&2&-2\\0&4&1\end{bmatrix}
$

* 消元法第2步操作  
$
E_{3,2}E_{2,1}A
=
\begin{bmatrix}1&0&0\\0&1&0\\0&-2&1\end{bmatrix}\begin{bmatrix}1&2&1\\0&2&-2\\0&4&1\end{bmatrix}=\begin{bmatrix}1&2&1\\0&2&-2\\0&0&5\end{bmatrix}
$

* 综合两步  
即$E_{32}(E_{21}A)=U$，也就是说如果我们想从$A$矩阵直接得到$U$矩阵的话，只需要$(E_{32}E_{21})A$即可。
